## (1)

```python
df_1['Join Date'] = pd.to_datetime(df_1['Join Date'])
df_1.head()
```

```python
df_1['Number_of_hours_per_day_in_self_study'] = df_1['Number of hours per year in self-study'] / 365
df_1.head()
```

```python
now = pd.to_datetime('2021-09-23')
df_1['Total_Days'] = (now - df_1['Join Date'])
df_1.head()
```

```python
df_1['Total_num_of_hrs_in_study'] = (df_1['Number_of_hours_per_day_in_self_study'] * df_1['Total_Days']).dt.days
df_1.head()
```

```python
start_date = '2010-01-01'
before_2010 = (df_1['Join Date'] < start_date)
```

```python
df_1.loc[before_2010].head()
```

```python
index = df_1.loc[before_2010].index
```

```python
df_1.loc[index, 'Total_num_of_hrs_in_study'] = df_1.loc[index, 'Total_num_of_hrs_in_study'] + 5
df_1.loc[index].head()
```

```python
senior = (df_1['Senior Management'] == True)
df_1.loc[senior].head()
```

```python
index2 = df_1.loc[senior].index
```

```python
df_1.loc[index2, 'Total_num_of_hrs_in_study'] = df_1.loc[index2, 'Total_num_of_hrs_in_study'] + 10
df_1.loc[index2].head()
```
